```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, accuracy_score

# 1. Memuat Data
file_path = 'earthquake_alert_balanced_dataset.csv'
df = pd.read_csv(file_path, delimiter=';')

print("--- Informasi Dataset Awal ---")
print(df.head())
print("\nJumlah baris dan kolom:", df.shape)
print("\nKolom target unik:", df['alert'].unique())

# 2. Pra-pemrosesan Data

# Mengubah kolom 'alert' (target) menjadi numerik
le = LabelEncoder()
df['alert_encoded'] = le.fit_transform(df['alert'])
# Mapping label:
# green -> 0, yellow -> 1, orange -> 2, red -> 3 (urutan tergantung abjad atau urutan kemunculan)
print("\nMapping Label Target:")
for i, label in enumerate(le.classes_):
    print(f"{label} -> {i}")

# Memisahkan Fitur (X) dan Target (y)
# Fitur: 'magnitude', 'depth', 'cdi', 'mmi', 'sig'
# Target: 'alert_encoded'
features = ['magnitude', 'depth', 'cdi', 'mmi', 'sig']
X = df[features]
y = df['alert_encoded']

# 3. Normalisasi Data Fitur (X)
# Normalisasi penting untuk ANN agar konvergensi lebih cepat dan stabil
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4. Membagi Data Training dan Testing
# 80% untuk training, 20% untuk testing
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print(f"\nUkuran Data Training: {X_train.shape}")
print(f"Ukuran Data Testing: {X_test.shape}")

# 5. Membuat dan Melatih Model ANN (MLPClassifier)
# MLP = Multi-layer Perceptron (salah satu jenis ANN)

# Hyperparameter Model:
# hidden_layer_sizes=(100, 50, 20): Tiga lapisan tersembunyi dengan 100, 50, dan 20 neuron
# activation='relu': Fungsi aktivasi Rectified Linear Unit
# solver='adam': Pengoptimal Adam
# max_iter=300: Jumlah maksimum iterasi/epoch

ann_model = MLPClassifier(
    hidden_layer_sizes=(100, 50, 20),
    activation='relu',
    solver='adam',
    max_iter=300,
    random_state=42,
    verbose=True # Tampilkan progress training
)

print("\n--- Mulai Melatih Model Artificial Neural Network (ANN) ---")
ann_model.fit(X_train, y_train)
print("--- Pelatihan Selesai ---")

# 6. Mengevaluasi Model
y_pred = ann_model.predict(X_test)

# Menghitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print(f"\n\n--- Hasil Evaluasi Model ANN ---")
print(f"Akurasi pada Testing Set: {accuracy*100:.2f}%")

# Laporan klasifikasi
print("\nLaporan Klasifikasi:")
# Menggunakan target_names dari LabelEncoder untuk tampilan yang lebih mudah dibaca
print(classification_report(y_test, y_pred, target_names=le.classes_))
```
